```python
import sqlite3
```

```python
conn = sqlite3.connect("house_price.db")
print(conn)
```

```python
cursor = conn.cursor()
```

```python
cursor.execute("""
    CREATE TABLE IF NOT EXISTS house_prices (
        price INTEGER PRIMARY KEY AUTOINCREMENT,
        bedrooms INTEGER,
        age INTEGER,
        area INTEGER
    )
""")
```

```python
cursor.execute('''
    INSERT INTO house_prices (price, bedrooms, age, area)
    VALUES
     (550000, 3.0, 20, 2600),
     (565000, 4.0, 15, 3000),
     (610000, 3.0, 18, 3200),
     (760000, 5.0, 8, 4000)
''')
```

```python
cursor.execute('SELECT * FROM house_prices ')
rows = cursor.fetchall()
# Display the fetched records
for row in rows:
    print(row)
```

```python
cursor.execute('UPDATE house_prices SET age = 10 WHERE price = 610000')
```

```python
cursor.execute('SELECT * FROM house_prices WHERE price = 610000')
updated_row = cursor.fetchone()
print("Updated Record:", updated_row)
```

```python
cursor.execute('DELETE FROM house_prices WHERE area = 2600')
```

```python
cursor.execute('SELECT * FROM house_prices')
remaining_rows = cursor.fetchall()
print("Remaining Records after Deletion:")
for row in remaining_rows:
    print(row)
```
